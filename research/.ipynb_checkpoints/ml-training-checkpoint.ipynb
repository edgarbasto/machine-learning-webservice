# Training ML algorithms

```python
import json
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #splitting data
from sklearn.preprocessing import LabelEncoder #preprocessing categorical
from sklearn.ensemble import RandomForestClassifier #training random forest
from sklearn.ensemble import ExtraTreesClassifier #training
import joblib #saving algorithm
```

```python
x_cols = [c for c in df.columns if c != 'income']
```

```python
x_cols
```

```python
X = df[x_cols]
y = df['income']
```

### Divide the data
70-30
random_state = 1234 (SEED)

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 1234)
```

```python
print(X_train.shape)
print(X_test.shape)
```

### Fill missing values

```python
#Get Mode for each variable with X_train.mode()
train_mode = dict(X_train.mode().iloc[0])
train_mode
```

```python
print(X_train.columns)
#Missing values overview
[X_train[i].isna().sum() for i in X_train.columns]
```

```python
#Total lines with missing values
X_train[X_train.isna().any(axis=1)]
```

```python
X_train = X_train.fillna(train_mode)
```

```python
#Validating 
X_train[X_train.isna().any(axis=1)]
```

### Convert Categoricals

```python
X_train.dtypes
```

```python
encoders = {}
for i in ['workclass', 'education', 'marital-status', 'occupation', 'relationship', 'race', 'sex','native-country']:
    convertCategorical = LabelEncoder()
    X_train[i] = convertCategorical.fit_transform(X_train[i])
    encoders[i] = convertCategorical
```

```python
encoders
```

```python
X_train.dtypes
```

# Algorithm training

#### Random Forest Algorithm

```python
rf = RandomForestClassifier(n_estimators= 100)
rf = rf.fit(X_train, y_train)
```

#### Extra Trees Algorithm

```python
et = ExtraTreesClassifier(n_estimators=100)
et = et.fit(X_train, y_train)
```
